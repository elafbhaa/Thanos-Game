```python
import os
import random
import time
```

```python
def show_universe_people(place):
    people = os.listdir(place)
    return people
```

```python
universe = show_universe_people('universe')
print(universe)
len_universe = len(universe)
print(len_universe)
```

```python
def thanos_snap(universe):
    half_of_universe = len(universe) // 2
    for i in range(half_of_universe):
        universe = show_universe_people('universe')
        random_killed = random.choice(universe)
        os.remove(f'universe/{random_killed}')
        time.sleep(.5)
        print(f"killed: {random_killed}")
    print('killed half of the universe')
```

```python
print(f" kille:{thanos_snap(universe)}")
```
